```python
delays_pred = delays_df.loc[:,['DEP_TIME', 'DEP_DEL15', 'DISTANCE', 'ARR_DEL15']]

delays_pred.head()
```

```python
X = delays_pred.loc[:,['DEP_TIME', 'DEP_DEL15', 'DISTANCE']]
y = delays_pred.loc[:,['ARR_DEL15']]
```

```python
X.head()
```

```python
y.head()
```

```python
sum(y.ARR_DEL15)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# train_test_split splits the data set into training and test datasets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
y_train.head()
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
# LinearRegression class fits a linear model, using the fit method

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# predict returns the values predicted for the test data by the trained model

y_pred = regressor.predict(X_test)
```

```python
y_test.head()
```

```python
y_pred
```

```python
# mean_squared_error will give the MSE

from sklearn import metrics
```

```python
metrics.mean_squared_error(y_test, y_pred)
```

```python
# calculate RMSE using numpy

import numpy as np

np.sqrt(metrics.mean_squared_error(y_test, y_pred))
```
