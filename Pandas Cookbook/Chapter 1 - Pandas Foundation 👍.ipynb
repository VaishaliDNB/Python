```python
"""Another option for debugging chains (other than stepwise incremental execution) is to call the .pipe method to show an 
intermediate value. The .pipe method on a Series needs to be passed a function that accepts a Series as input and can 
return anything (but we want to return a Series if we want to use it in a method chain)."""

def debug_ser(ser):
    print ("BEFORE")
    print (ser)
    print ("AFTER")
    return (ser)
```

```python
col_map = {
    "director_name" : "director",
    "num_critic_for_reviews" : "critic_reviews"
}
movie.rename(columns=col_map)
```

```python
movie.head(3).T
```

```python
# renaming index - setting index to movie title column and changing some index values

idx_map = {
    "Avatar" : "Rattva",
    "Spectre" : "Ertceps",
    "Pirates of the Caribbean: At World's End" : "POC"
}

col_map = {
    "aspect_ratio" : "aspect",
    "movie_facebook_likes" : "fblikes"
}


(
    movie.set_index("movie_title")
    .rename(index = idx_map, columns = col_map)
    .head(5)
)
```

```python
# Option 2 - reassign the index and column attributes to a Python list, make needed changes and assign to index and column names
# Option 3 - read a column as the index

movies = pd.read_csv("D:\Learning Data Science\Python - Pandas\Pandas cookbook\Pandas-Cookbook-master\data\movie.csv",
                     index_col = "movie_title") # reading data as movies
ids = movies.index.to_list()
columns = movies.columns.to_list()
ids[1] = "POC"
columns[1] = "director"
columns[-1] = "fblikes"
movies.index = ids
movies.columns = columns
movies.head(3)
```

```python
# Option 4 - pass a function in the rename method, e.g. to clean up column names

def to_clean(val):
    return val.strip().lower().replace(" ","_")

movies.rename(columns=to_clean).head(3)
```

```python
# Option 5 - use list comprehensions to clean up column names

cols = [col.strip().lower().replace(" ","_") for col in movies.columns]
movies.columns = cols
movies.head()
```

### Creating and deleting columns

```python
# Option 1 - good for chaining
movies = pd.read_csv("D:\Learning Data Science\Python - Pandas\Pandas cookbook\Pandas-Cookbook-master\data\movie.csv")
movies["has_seen"] = 0
movies.head(3)
```

```python
# Option 2 - assign method
movies = pd.read_csv("D:\Learning Data Science\Python - Pandas\Pandas cookbook\Pandas-Cookbook-master\data\movie.csv")
movies.assign(has_seen=0)
```

```python
# one way to add existing column values
total = movies.actor_1_facebook_likes + movies.actor_2_facebook_likes + movies.actor_3_facebook_likes + movie.director_facebook_likes

total.head()
```

```python
# another way to add existing column values
cols = [
    "actor_1_facebook_likes",
    "actor_2_facebook_likes",
    "actor_3_facebook_likes",
    "director_facebook_likes"
]
sum_col = movies.loc[:,cols].sum(axis="columns")
sum_col
```

```python
movies.assign(total_likes = sum_col).head(5)
```

```python
# pass in a function as the value of the parameter in the call to the .assign method
def sum_likes(df):
    return df[
        [
            c
            for c in df.columns
            if "like" in c
            and ("actor" in c or "director" in c)
        ]
    ].sum(axis=1)
movies.assign(total_likes=sum_likes).head(3).T
```

```python
# comparing total and sum_col
(
    movies.assign(total_likes = total)["total_likes"]
    .isna()
    .sum()
)
```

```python
(
    movies.assign(total_likes = sum_col)["total_likes"]
    .isna()
    .sum()
)
```

```python
# filling the missing values with 0
(
    movies.assign(total_likes = total.fillna(0))["total_likes"]
    .isna()
    .sum()
)
```

```python
# checking if cast_total_facebook_likes is greater than or equal to total_likes

def cast_like_gt_actor (df):
    return (df["cast_total_facebook_likes"] >= df["total_likes"] )

df2 = movies.assign(total_likes=total,is_cast_likes_more=cast_like_gt_actor)

df2["is_cast_likes_more"].all()

# df2 = movies.assign(total_likes = total, is_cast_likes_more = (movies.cast_total_facebook_likes>=total))
# df2
```

```python
#  drop column

df2 = df2.drop(columns="total_likes")
```

```python
actor_sum = movies[
    [
        c
        for c in movies.columns
        if "actor_" in c and "_likes" in c
    ]
].sum(axis = "columns")

actor_sum
```

```python
# calculate the percentage of the cast_total_facebook_likes that come from actor_sum

pct_like = actor_sum.div(movies["cast_total_facebook_likes"]).mul(100)

pct_like.describe()
```

```python
pd.Series(pct_like.to_numpy(), index=movies["movie_title"]).head()
```

```python
movies.insert(loc=profit_index, column = "profit", value = movies["gross"] - movies["budget"])
movies.head(5)
```
