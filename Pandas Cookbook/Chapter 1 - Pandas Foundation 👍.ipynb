### Chaining Series Methods

```python
# In Python, many attributes and methods return new objects. 
# This allows sequential invocation of methods using attribute access. This is called method chaining or flow programming.
```

```python
director.value_counts().head(3)
```

```python
fb_likes.isna().sum()      #counting the number of missing values
```

```python
fb_likes.dtype
```

```python
fb_likes.fillna(0).astype(int).head()
```

```python
"""Another option for debugging chains (other than stepwise incremental execution) is to call the .pipe method to show an 
intermediate value. The .pipe method on a Series needs to be passed a function that accepts a Series as input and can 
return anything (but we want to return a Series if we want to use it in a method chain)."""

def debug_ser(ser):
    print ("BEFORE")
    print (ser)
    print ("AFTER")
    return (ser)

fb_likes.fillna(0).pipe(debug_ser).astype(int).pipe(debug_ser).head()
```

```python
# creating a global variable to store an intermediate value using .pipe

intermediate = None
def get_intermediate(ser):
    global intermediate
    intermediate = ser
    return ser
```

```python
res = fb_likes.fillna(0).pipe(get_intermediate).astype(int).head()
res
```

```python
intermediate
```

```python
(
    fb_likes
    .fillna(0)
    .astype(int)
    .head()
)
```

```python
# Option 1 - index assignment - this is not good for chaining methods
movies = pd.read_csv("D:\Learning Data Science\Python - Pandas\Pandas cookbook\Pandas-Cookbook-master\data\movie.csv")
movies["has_seen"] = 0
movies.head(3)
```

```python
# Option 2 - assign method - this is good for chaining methods
movies = pd.read_csv("D:\Learning Data Science\Python - Pandas\Pandas cookbook\Pandas-Cookbook-master\data\movie.csv")

idx_map = {"Avatar" : "Rattva",
           "Spectre" : "Ertceps",
           "Pirates of the Caribbean: At World's End" : "POC"}
col_map = {"aspect_ratio" : "aspect",
           "movie_facebook_likes" : "fblikes"}

(movie.set_index("movie_title")
 .rename(index = idx_map, columns = col_map)
 .assign(has_seen = 0)
 .head(5))
```

```python
# one way to add existing column values
total = (movies.actor_1_facebook_likes 
         + movies.actor_2_facebook_likes 
         + movies.actor_3_facebook_likes 
         + movie.director_facebook_likes)
total.head()
```

```python
movies.assign(total_likes = total).head(5)
```

```python
# another way to add existing column values
cols = ["actor_1_facebook_likes",
        "actor_2_facebook_likes",
        "actor_3_facebook_likes",
        "director_facebook_likes"]
sum_col = movies.loc[:,cols].sum(axis="columns")
sum_col
```

```python
movies.assign(total_likes = sum_col).head(5)
```

```python
# one more way is to pass in a function as the value of the parameter in the call to the .assign method
def sum_likes(df):
    return df[
        [
            c
            for c in df.columns
            if "like" in c
            and ("actor" in c or "director" in c)
        ]
    ].sum(axis=1)
movies.assign(total_likes=sum_likes).head()
```

```python
# comparing total and sum_col - using the + operator, the result had NaNs, but .sum ignores missing numbers by default
movies.assign(total_likes = total)["total_likes"].isna().sum()
```

```python
movies.assign(total_likes = sum_col)["total_likes"].isna().sum()
```

```python
# filling the missing values in total with 0
movies.assign(total_likes = total.fillna(0))["total_likes"].isna().sum()
```

```python
# checking if cast_total_facebook_likes is greater than or equal to total_likes

def cast_like_gt_actor (df):
    return (df["cast_total_facebook_likes"] >= df["total_likes"] )

df2 = movies.assign(total_likes=total,is_cast_likes_more=cast_like_gt_actor)

df2["is_cast_likes_more"].all()

# df2 = movies.assign(total_likes = total, is_cast_likes_more = (movies.cast_total_facebook_likes>=total))
# df2.head()
```

```python
# .drop method defaults to dropping rows by the index names. The default value for axis is 0 or 'index'.
# To drop columns, you must set the axis parameter to either 1 or 'columns'. 

df2 = df2.drop(columns="total_likes")
```
