### Changing datatype of a column to reduce memory usage

```python
different_cols = ["RELAFFIL","SATMTMID","CURROPER","INSTNM","STABBR",]

col2 = college.loc[:,different_cols]

col2
```

```python
col2.dtypes
```

```python
orig_mem = col2.memory_usage(deep=True) # deep = True is needed for the correct data size of the object columns
orig_mem
```

```python
col2.select_dtypes(include=["number"]).nunique()
```

```python
# converting RELAFFIL AND CURROPER, both containing only 0 and 1, to an 8 bit integer

col2["RELAFFIL"] = col2["RELAFFIL"].astype(np.int8)
col2["CURROPER"] = col2["CURROPER"].astype(np.int8)
```

```python
col2.dtypes
```

```python
col2.memory_usage(deep=True)
```

```python
col2.select_dtypes(include=["object"]).nunique()
```

```python
# converting STABBR, containing 59 values, to a categorical type

col2["STABBR"] = col2["STABBR"].astype("category")
```

```python
col2.dtypes
```

```python
new_mem = col2.memory_usage(deep=True)
new_mem
```

```python
new_mem/orig_mem
```

### Selecting the smallest of the largest

```python
# nlargest and nsmallest

movie = pd.read_csv('D:\Learning Data Science\Python - Pandas\Pandas cookbook\Pandas-Cookbook-master\data\movie.csv')
```

```python
movie.columns
```

```python
movie2 = movie[['movie_title', 'imdb_score', 'budget']]
```

```python
movie2.head()
```

```python
movie2.nlargest(100, 'imdb_score')
```

```python
movie2.nlargest(100, 'imdb_score').nsmallest(5, 'budget')
```

### Selecting the largest of each group by sorting

```python
movie.columns
```

```python
movie[['movie_title','title_year','imdb_score']]
```

```python
movie[['movie_title','title_year','imdb_score']].sort_values('title_year', ascending = True)
```

```python
movie[['movie_title','title_year','imdb_score']].sort_values(['title_year', 'imdb_score'], ascending = False)
```

```python
movie[['movie_title','title_year','imdb_score']].sort_values(['title_year', 'imdb_score'], ascending = False).drop_duplicates(subset = 'title_year')
```

```python
# using groupby for the same

(
movie[['movie_title', 'title_year', 'imdb_score']]
.groupby("title_year", as_index = False)
.apply(lambda df: df.sort_values('imdb_score', ascending = False).head(1))
.droplevel(0)
.sort_values('title_year', ascending = False)
)
```

```python
# sorting one column in ascending order and the other in descending order

(
movie[['movie_title', 'title_year', 'content_rating', 'budget']]
.sort_values(['title_year', 'content_rating', 'budget'], ascending = [False, False, True])
.drop_duplicates(subset = ['title_year', 'content_rating'])
)
```

### Replicating nlargest with sort_values

```python
# Selecting the smallest of the largest

(
movie[['movie_title', 'imdb_score', 'budget']]
.nlargest(100, 'imdb_score')
.nsmallest(5, 'budget')
)
```

```python
# replicating the above using sort_values

(
movie[['movie_title', 'imdb_score', 'budget']]
.sort_values('imdb_score', ascending = False)
.head(100)
.sort_values('budget')
.head(5)
)
```

### Calculating a trailing stop order price

```python
import datetime
import pandas_datareader.data as web
import requests_cache

session = requests_cache.CachedSession(cache_name = 'cache', backend = 'sqlite', expire_after = datetime.timedelta(days=90))

# getting the Tesla Motors (TSLA) stock price
tsla = web.DataReader('tsla', data_source='yahoo', start = '2017-1-1', session = session)
```

```python
tsla.head(8)
```

```python
tsla
```

```python
tsla_close = tsla["Close"]
```

```python
tsla_cummax = tsla_close.cummax()
tsla_cummax.head()
```

```python
tsla_cummax.mul(0.9).head()
```

```python
# chaining all above together

tsla['Close'].cummax().mul(0.9).head()
```
